```python
import bmll2 as b2
from bmll2 import reference, Security, NormalisedSecurity, SparkHelper, get_market_data, get_market_data_range, VenueMarketError, get_market_tables, save_spark_dataframe, load_spark_dataframe

import random
import math
import pandas as pd
import polars as pl
import numpy as np
from pandas import StringDtype
import gc

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.ticker import LogFormatterSciNotation
from scipy.stats import linregress
from scipy.optimize import curve_fit
from scipy.stats import t
```

```python
N                   = 20
trader_distribution = 'power'
alpha               = 2
group               = 'day'
identifier          = f'{trader_distribution}_{N}_{group}_pl'
exchange            = 'XJSE'
ticker              = 'GRT'
```

```python
b2.get_file(f'data/impact_decay/{exchange}/{identifier}/{ticker}_{exchange}_impact_decay_{identifier}.parquet')
decay_data = pl.read_parquet(f'{ticker}_{exchange}_impact_decay_{identifier}.parquet')
```

```python
decay_data.head()
```

```python
bins = np.linspace(1, 2, 60)

df = decay_data.with_columns(pl.col('z').cut(bins).alias('bin'))
df = df.group_by('bin').agg(pl.col('scaled impact').mean().alias('scaled impact (binned)')).sort('bin')

bin_mids = (bins[:-1] + bins[1:]) / 2
```

```python
df
```

```python
def decay_model(z, beta, Y):
    return  Y * (z ** (1 - beta) - (z - 1) ** (1 - beta))
    #return Y * (z ** beta)

z      = np.array(bin_mids)
impact = np.array(df['scaled impact (binned)'][1: ])

# remove first point if z=1 exactly (can cause instability)
mask  = (z > 1) & (impact > 0)
z_fit = z[mask]
impact_fit = impact[mask]

params, covariance = curve_fit(decay_model, z_fit, impact_fit, p0 = [0.2, 0.5], maxfev = 10000)

beta_est = params[0]
I_est    = params[1]

print('Estimated beta:', beta_est)
print('Estimated Y:', I_est)


fitted_curve = decay_model(z_fit, beta_est, I_est)

n   = len(z_fit)
p   = len(params)          # number of parameters (2 here)
dof = n - p                # degrees of freedom

alpha = 0.05               # 95% CI
tval  = t.ppf(1 - alpha / 2, dof)
se    = np.sqrt(np.diag(covariance))
interval_half_widths = tval * se[0], tval * se[1]
print('Interval half widths:', interval_half_widths)

Y_ci        = I_est - tval * se[1], I_est + tval * se[1]
exponent_ci = beta_est - tval * se[0], beta_est + tval * se[0]
print('Scaling Y 95% CI:', Y_ci)
print('Exponent 95% CI:', exponent_ci)
```
